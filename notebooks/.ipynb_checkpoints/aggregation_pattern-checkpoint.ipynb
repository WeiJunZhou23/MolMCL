```python
import torch
import yaml
from rdkit import Chem
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import colormaps

from molmcl.finetune.loader import MoleculeDataset
from molmcl.finetune.model import GNNPredictor
from molmcl.utils.draw import draw_mols, draw_mol_with_highlight
from molmcl.utils.data import get_scaffold, generate_motif_features
```

```python
# Setup config (select molecules in bbbp as an example)
with open('../config/moleculenet/bbbp.yaml', 'r') as f:
    config = yaml.load(f, Loader=yaml.FullLoader)
config['model']['checkpoint'] = '../checkpoint/zinc-gps_best.pt'
config['model']['backbone'] = 'gps'
config['dataset']['feat_type'] = 'super_rich'
config['dataset']['data_dir'] = '../data/finetune/moleculenet'
config
```

```python
# Setup dataset
dataset = MoleculeDataset(config['dataset']['data_dir'],
                          config['dataset']['data_name'],
                          config['dataset']['feat_type'])

# Setup model
model = GNNPredictor(num_layer=config['model']['num_layer'],
                     emb_dim=config['model']['emb_dim'],
                     normalize=config['model']['normalize'],
                     num_tasks=1, atom_feat_dim=170, bond_feat_dim=14,
                     use_prompt=config['model']['use_prompt'],
                     drop_ratio=config['model']['dropout_ratio'],
                     temperature=config['model']['temperature'],
                     model_head=config['model']['heads'], aggr_head=4, 
                     layer_norm_out=True, backbone='gps')

loaded_state_dict = torch.load(config['model']['checkpoint'])['wrapper']
model.load_state_dict(loaded_state_dict, strict=False)
print('Checkpoint loaded.')
```

```python
# Choose a random molecule from dataset
sample = dataset[np.random.randint(len(dataset))]  # choose by index
print(sample.smi)
```

```python
# Compute the attention score
with torch.no_grad():
    graph_rep = model.get_representations(sample, return_score=True)
```

```python
# Visualization of scaffold and functional groups
size = (300, 150)
mol = Chem.MolFromSmiles(sample.smi)
draw_mols([sample.smi], size=size, straighten=True)
draw_mols([sample.smi], [get_scaffold(sample.smi)], size=size, straighten=True)  # highlighted scaffold

matched_frags, smarts_list = generate_motif_features(mol, verbose=False, return_name=True) 
selected_names = []
selected_smarts_list = []
for i, (name, smarts) in enumerate(smarts_list):
    print(name)
    if not selected_names or name in selected_names:
        selected_smarts_list.append(smarts)
        color = list(colormaps['Paired'](i))
        draw_mols([sample.smi], [smarts], size=size, color=matplotlib.colors.to_rgb(color), straighten=True)
```

```python
# Visualize the aggregation patterns (node attention)
factor = 5  # change the factor for better illustration
for p_i in range(3):
    if p_i < 2:
        print('Attention #{}.1:'.format(p_i+1))
        atom_scores = graph_rep[1][p_i][0, 0, 0].squeeze(0)
        atom_scores = atom_scores[:len(mol.GetAtoms())]
        
        plt.figure(figsize=(5, 1.3))
        for i, spine in enumerate(plt.gca().spines.values()):
            if i in [0, 1, 3]:
                spine.set_visible(False)
        plt.bar(np.arange(len(atom_scores)), atom_scores)
        plt.xticks(np.arange(len(atom_scores)), 
                   [atom.GetSmarts() for atom in mol.GetAtoms()], 
                   rotation=90, fontsize=10)
        plt.tick_params(top=False, bottom=True, left=False, right=False, labelleft=False, labelbottom=True)
        plt.show()
        draw_mol_with_highlight(mol, atom_scores.numpy()*factor, size=size, straighten=True)
    else:
        for j in range(4):
            print('Attention #{}.{}:'.format(p_i+1, j+1))
            atom_scores = graph_rep[1][p_i][0, j, 0].squeeze(0)
            atom_scores = atom_scores[:len(mol.GetAtoms())]
            
            plt.figure(figsize=(5, 1.3))
            for i, spine in enumerate(plt.gca().spines.values()):
                if i in [0, 1, 3]:
                    spine.set_visible(False)
            plt.bar(np.arange(len(atom_scores)), atom_scores)
            plt.xticks(np.arange(len(atom_scores)), 
                       [atom.GetSmarts() for atom in mol.GetAtoms()], 
                       rotation=90, fontsize=10)
            plt.tick_params(top=False, bottom=True, left=False, right=False, labelleft=False, labelbottom=True)
            plt.show()
            draw_mol_with_highlight(mol, atom_scores.numpy()*factor, size=size, straighten=True)    
```
